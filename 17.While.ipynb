# While
* While in Phthon language is used to execute statements as long as the given condition is met (True)
* The condition can be any expression and please note that True in Python includes all non-zero values
* When the condition becomes False, the program will continue to the line after the statement block

Be careful, later the output will appear continuously without stopping:

```python
# If a number less than 3 is entered, 
the output will appear non-stop
number = int(input())

while (number< 3):
	print("numeral")
```

```python
number = int(input())

while (number<3) :
    print("number {}".format(number))
    number = number + 1
```

The difference in the placement of the "number = number+2" syntax, can cause different outputs such as the 2 syntaxes and the output below:

```python
number = int(input())

while (number<10):
    number = number + 2
    print("number {}".format(number))
```

```python
number = int(input())

while (number<10):
    print("number {}".format(number))
    number = number + 2
```

# Else after while

```python
number = int(input("Want to display a number up to minus how many start the number with a sign - eg (-11):"))
i = 0

while (i > number):
    i = i - 1
    print("number {}".format(i))
```
